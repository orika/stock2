#### 两只股票的相关性分析

```python
import tushare as ts
import pandas as pd
import seaborn as sns

ts.set_token('d3fdbde82434cd6d7897550852136449f9fcba912e3eacb47b004600')
pro = ts.pro_api()


pa_price = ts.pro_bar(ts_code='000001.SZ', adj='qfq', start_date='20180110', end_date='20190625')

pa_pct_chg = pa_price['pct_chg'].dropna()


hk_price = ts.pro_bar(ts_code='002415.SZ', adj='qfq', start_date='20180110', end_date='20190625')

hk_pct_chg = hk_price['pct_chg'].dropna()


sns.jointplot(pa_pct_chg, hk_pct_chg, kind='reg', height=12)
```
